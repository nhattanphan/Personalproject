```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fifa_data = pd.read_csv("data.csv")
fifa_data
```

# Question 1
Overall of all players

```python
plt.hist(fifa_data['Overall'],10, color='blue') 
plt.show()
```

# Question 2
Average age of all players

```python
plt.hist(fifa_data['Age'], color='blue')
plt.show()
```

# Question 3
Which players have overall higher than 90 ?

```python
fifa_data[fifa_data['Overall'] > 90][['Name','Overall',"Club","Nationality"]].head(20)
```

# Question 4
Number of players by positon

```python
position_count = fifa_data['Position'].value_counts()
plt.figure(figsize=(20,15))
sns.barplot(position_count.values,position_count.index,)
plt.title('Number of players by position', fontsize=20)
plt.ylabel('Position', fontsize=15)
plt.xlabel('Players', fontsize=15)
plt.show()
```

# Question 5
Number of players by nationality

```python
nation_count = fifa_data['Nationality'].value_counts()
nation_count = nation_count[0:20]
plt.figure(figsize=(20,15))
sns.barplot(nation_count.values,nation_count.index, )
plt.title('Number of players')
plt.ylabel('Nationality', fontsize=12)
plt.xlabel('Players', fontsize=12)
plt.show()
```

# Question 6
Count of players on Basis of Height

```python
plt.figure(figsize = (13, 8))
sns.countplot(x = 'Height', data = fifa_data, palette = 'dark')
plt.title(label = 'Count of players on Basis of Height', fontsize = 20)
plt.xlabel(xlabel = 'Height in Foot per inch', fontsize = 16)
plt.ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```

# Question 7
Preferred Foot

```python
preferred_foot = fifa_data.loc[:,'Preferred Foot'].value_counts()
preferred_foot.plot.bar(rot=40,color="y")
plt.title('Preferred Foot of Players')
plt.ylabel('Numbers', fontsize=15)
plt.xlabel('Foot', fontsize=15)
```

# Question 8
Top 10 right-footed players

```python
fifa_data[fifa_data['Preferred Foot'] == 'Right'][['Name','Overall',"Club","Nationality"]].head(10)
```

# Question 9
Overall by Clubs

```python
plt.figure(figsize = (10,5))
pl = sns.boxplot(x=fifa_data['Club'][1:50], y=fifa_data['Overall'], palette='hls');
pl.set_title(label='Overall - Club');
plt.xlabel('Club')
plt.ylabel('Overall')
plt.xticks(rotation = 80)
plt.ylim(85, 95)
plt.show()
```
